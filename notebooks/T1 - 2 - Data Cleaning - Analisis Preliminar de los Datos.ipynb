## Datos/Valores perdidos o faltantes

```python
pd.isnull(data["body"])
#Para ver si faltan valores en la columna body
```

```python
pd.notnull(data["body"])
#El efecto contrario al anterior
```

```python
pd.isnull(data["body"]).values.ravel().sum()
#Nos convierte los valores en un array solo values, ravel crea un unico array de datos porque con solo value se 
#pueden duplicar y sum suma los trues, es decir, los valores faltantes
```

```python
pd.notnull(data["body"]).values.ravel().sum()
#Aqui ocurre el efecto contratio, se suman los valors true, es decir los que si tienen valores
```

Los valores que faltan en un dataset pueden venir por dos razones:
* Extraccion de los datos: incompatibilidades entre el origen y el destino del dataset
* Recolección de los datos

#### Borrado de valores que faltan 

```python
data.dropna(axis = 0, how="all")
# axis = 1 borra toda la columna
# axis = 0 borra toda la fila y how = "all" borra toda la fila si, todas las columnas están vacías
```

```python
data2 = data
```

```python
data.dropna(axis=0, how="any")
# Borrar cualquier columna que tenga al menos un valor vacio o NA
```

```python
# Reemplazar los valores faltantes por otros
data3 = data
```

```python
data3.fillna(0)
```

```python
data4 = data
```

```python
data4.fillna("desconocido")
```

```python
data5 = data
```

```python
data5 ["body"] = data5["body"].fillna(0)
# Solo la columna body se remplaza los valores por 0
data5["home.dest"] = data5["home.dest"].fillna("desconocido")
# Se pone data5 = para sobre escribir el dataset
data5.head(5)
```

```python
pd.isnull(data5["age"]).values.ravel().sum()
# Los valores vacios de la columna age se suman
```

```python
data5["age"].fillna(data["age"].mean())
# Los valores vacios de la columna age, se reemplazan por "mean" el promedio
```

```python
data5["age"][1291]
```

```python
data5["age"].fillna(method="ffill")
# fillna rellena el valor vacio o faltante con el metodo ffill que significa el valor conocido adelante o forward
# o sea poner en los valores anteriores al valor conocido (que está adelante), ese valor
```

```python
data5["age"].fillna(method="bfill")
# Poner el valor conocido en los valores vacios que están delante de él, o sea el valor conocido está atrás (Back)
```
